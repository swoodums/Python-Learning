```python
h = pd.read_csv("california_housing_test.csv")  #loads the csv data into pandas DataFrame "h"
```

#How do I select a subset of a **DataFrame**?
#How do I select specific columns from a **DataFrame**?

I'm interested in the median ages column.

```python
ages = h["housing_median_age"]  #To select a single column, use square brackets with the column name
ages.head()  #previews column
```

```python
h["housing_median_age"].shape  #inspects the shape of the dataframe or column (number of rows, number of colums)  .shape is an attribute and needs no brackets like a method/function.
```

```python
h.shape #inspects the shape of the whole dataframe
```

I'm interested in the median age and the population.

```python
age_pop = h[["housing_median_age", "population"]]  #selecting multiple columns using a list of column names.
                                                   #Note that the inner square brackets define a python list, whereas the outer brackets are used to select the data from the dataframe.
                                                   #The returned data type is a DataFrame
age_pop.head()  #preview data
```

```python
type(age_pop)   #The returned data type is a Dataframe
```

```python
age_pop.shape
```

#How do I filter specific rows from a DataFrame?
I'm interested in areas where the median age is less than 35.
To select rows based on a conditional expression, use a condition inside the selection brackets []

```python
under_35 = h[h['housing_median_age'] < 35]  #The condition inside the selection brackets h['housing_media_age'] < 35 checks for which rows the age column has a value less than than 35.
under_35  #previews data
```

```python
under_35.shape  #we can see that we have 1066 fewer rows in this dataframe.
```

#How do I select specific rows and columns from a DataFrame?
I'm interested in the population for areas where the median age is under 35.

```python
young_population = h.loc[h['housing_median_age'] < 35, 'population']  #This uses the loc operator, which selects both columns and rows from the DataFrame.
young_population
```

```python
h.iloc[9:25, 2:5] #iloc can also be used to specify column and rows by their index rather than their labels.
```

```python
#When selecting specific rows and/or columns with iloc, new values can be assigned to the selected data.
h.iloc[0,0] = -118.07 #This reassigned the first row of the first column with a new longitude of -118.07
h
```

#RECAP


*   When selecting subsets of data, square brackets [ ] are used.
*   Inside square brackets, you can use a single column/row label, a list of column/row labels, a slice of labels, a conditional expression, or a colon.
*   Select specific rows and/or columns using loc when using the row and column names.
*   Select specific rows and/or columns using iloc when using the row and column indices (location of data)
*   You can assign new values to a selection based on loc and iloc.
