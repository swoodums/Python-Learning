#How to create new columns derived from existing columns?

I want to express the N02 concentration of the station in London as mg/m^3.

```python
air_quality = pd.read_csv('https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/air_quality_no2.csv', index_col=0, parse_dates=True)
```

```python
air_quality['london_mg_per_cubic'] = air_quality['station_london'] * 1.882  # To create a new column, use the [] brackets with the new column name at the left side of the assignment.  The 1.882 is the conversion factor.  The calculation is done element_wise, meaning it applies to each row in a column all at once.
air_quality.head
```

I want to check the ratio of the values in Paris versus Antwerp and save the result in a new column.

```python
air_quality['ratio_paris_antwerp'] = air_quality['station_paris'] / air_quality['station_antwerp']  # Any mathematical operators and logical operators also work element-wise (i.e. work on the whole column)
air_quality.head()
```

I want rename the data columns to the corresponding station identifiers used by OpenAQ.

```python
air_quality_renamed = air_quality.rename(  #the rename() function can be used for both column labels and row labels.
    columns={
        'station_antwerp': 'BETR801',
        'station_paris': 'FR04014',
        'station_london': 'London Westminster',
    }
)
air_quality_renamed.head()
```

```python
air_quality_renamed = air_quality_renamed.rename(columns = str.lower) #rename can also take functions and apply them to the columns or rows.  Here, we are renaming our columns to be the lowercase version of the existing label.
air_quality_renamed.head()
```

#Summary and Recap

*   Create a new column by assigning the output to the DataFrame with a new column name in between the [ ].
*   Operations are element-wise, no need to loop over rows.
*   Use rename with a dictionary or function to rename row labels or column names.
